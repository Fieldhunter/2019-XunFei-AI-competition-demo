```python
import json
import pandas as pd
```

```python
data1 = pd.read_csv("~/桌面/移动广告/data/train.csv")
data2 = pd.read_csv("~/桌面/移动广告/data/test.csv")
```

# 对以下七个字段进行索引

```python
col_name = ["adunitshowid","mediashowid","apptype","city","carrier","osv","lan"]
index_dict = {}
```

```python
data1
```

```python
for i in col_name:
    all_set = set(data1[i]) & set(data2[i])
    index_dict[i] = list(all_set)
```

```python
index_dict
```

```python
len(index_dict)
```

```python
with open("第一轮字段索引.json","w") as f:
    json.dump(index_dict,f)
```
