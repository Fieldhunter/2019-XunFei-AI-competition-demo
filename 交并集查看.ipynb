```python
col_name1 = data1.columns
col_name2 = data2.columns
```

```python
for i in col_name2:
    if i == "label":
        continue
    try:
        sid1 = set([item.lower() for item in list(data1[i])])
        sid2 = set([item.lower() for item in list(data2[i])])
    except:
        sid1 = set(data1[i])
        sid2 = set(data2[i])
    print("{}: 交集---{}，并集---{}".format(i,len(sid1&sid2),len(sid1|sid2)))
```
